```python
# the problem is that this is a string we need to convert this into actual format
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def extract_genres(genres_dict):
    list = []
    for i in ast.literal_eval(genres_dict):
        list.append(i['name'])
    return list
```

```python
df['genres'] = df['genres'].apply(extract_genres)
```

```python
df['keywords'] = df['keywords'].apply(extract_genres)
```

```python
df.head()
```

```python
def extract1(genres_dict):
    list = []
    counter = 0
    for i in ast.literal_eval(genres_dict):
        if counter == 3:
            break
        counter = counter + 1
        list.append(i['name'])
    return list
```

```python
df['cast'] = df['cast'].apply(extract1)
```

```python
def extract_director_name(genres_dict):
    list = []
    for i in ast.literal_eval(genres_dict):
        if i['job'] == 'Director':
            list.append(i['name'])
            break
    return list
```

```python
df['crew'] = df['crew'].apply(extract_director_name)
```

```python
df['overview'] = df['overview'].apply(lambda x : x.split())
```

```python
df.head(2)
```

```python
# we don't want this much columns but want to create tags

# and also we want tags so need to create a name with no spaces
#  ex :- SalmaanKhan, ScienceFiction etc
```

```python
df['genres'] = df['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
df['keywords'] = df['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
df['cast'] = df['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
df['crew'] = df['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
df
```

```python
df['tag'] = df['overview'] + df['cast'] + df['crew'] + df['keywords'] + df['genres']
df['tag'][0]
```

```python
df = df[['id', 'title' , 'tag']]
df.head(2)
```

```python
df['tag'] = df['tag'].apply(lambda x : " ".join(x))
```

```python
# lower everything
df['tag'] = df['tag'].apply(lambda x : x.lower())
```

```python
# we need to do lemitization as 'action' and 'actions' are same
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
ps.stem('loving')
```

```python
# stemming
def stem(text):
    list = []
    for word in text.split():
        list.append(ps.stem(word))

    return " ".join(list)
```

```python
df['tag'] = df['tag'].apply(stem)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english', max_features=2000)
vectors = tfidf.fit_transform(df['tag']).toarray()
vectors
```

```python
tfidf.get_feature_names_out()
```

```python
# get the cosine similaritites of one vector to all others
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
# to get the movie with most similarity score
sorted(list(enumerate(similarity[0])), reverse=True, key = lambda x : x[1])
```

```python
def recommend(movie):
    movie_index = df[df['title'] == movie].index[0]

    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key = lambda x : x[1])[1:6]

    for i in movies_list:
        print(df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
